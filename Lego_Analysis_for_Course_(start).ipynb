### Aggregate Data with the Python .agg() Function

Let's work out the number of different themes shipped by year. This means we have to count the number of unique theme_ids per calendar year.

```python
themes_by_year = sets.groupby("year").agg({'theme_id': pd.Series.nunique}) # count the number of unique theme_ids per calendar year

themes_by_year.rename(columns= {"theme_id" : "nr_themes"}, inplace=True)

themes_by_year.tail()
```

**Challenge**: Plot the number of themes released by year on a line chart. Only include the full calendar years (i.e., exclude 2020 and 2021).

```python
plt.xticks(fontsize=12)
plt.yticks(fontsize=14)
plt.xlabel("Year", fontsize=14)
plt.ylabel("Number of themes", fontsize=14)
plt.plot(themes_by_year.index[:-2], themes_by_year["nr_themes"][:-2])
```

### Line Charts with Two Seperate Axes

```python
ax1 = plt.gca() # get current axis

ax2 = ax1.twinx() # create another exis and allow them to share the same x-axis

ax1.set_xlabel("Year", fontsize=14)

ax1.set_ylabel("Number of sets", fontsize=12, color='green')
ax2.set_ylabel("Number of themes", fontsize=12, color='blue')

ax1.plot(sets_by_year.index[:-2], sets_by_year["set_num"][:-2], color='green')
ax2.plot(themes_by_year.index[:-2], themes_by_year["nr_themes"][:-2])
```

**Challenge**: Use the <code>.groupby()</code> and <code>.agg()</code> function together to figure out the average number of parts per set. How many parts did the average LEGO set released in 1954 compared to say, 2017?

```python
parts_per_set = sets.groupby("year").agg({"num_parts" : pd.Series.mean})
```

### Merging (i.e., Combining) DataFrames based on a Key


```python
# adding column headings to our set_theme_count df
set_theme_count = pd.DataFrame({
    "id" : set_theme_count.index,
    "set_count" : set_theme_count.values,
})
set_theme_count
```

```python
merged_df = pd.merge(set_theme_count, themes, on='id') # to merge 2 dataframes based on a commonkey

merged_df.head()
```

```python
plt.figure(figsize=(14, 8))
plt.xticks(fontsize=12, rotation=45)
plt.yticks(fontsize=14)
plt.xlabel("Theme name", fontsize=14)
plt.ylabel("Num of sets", fontsize=14)
plt.bar(merged_df["name"][:10], merged_df["set_count"][:10])
```
